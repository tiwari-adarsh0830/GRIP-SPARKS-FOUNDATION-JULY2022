### Handling the Null Values 

    We will try to fill null values of continuous variables by median values and classification variables by mode values.

```python
#these are the colums having null values , we try to finde their mode and fill the null values with them
print(df_terror["state"].mode(),df_terror["city"].mode(),df_terror["targsubtype"].mode(),df_terror["target1"].mode(),df_terror["motive"].mode(),df_terror['weaponsub'].mode())
```

```python
df_terror["state"]=df_terror["state"].fillna("Baghad")
df_terror["city"]=df_terror["city"].fillna("Unknown")
df_terror["targsubtype"]=df_terror["targsubtype"].fillna("Unnamed Civilian/Unspecified")
df_terror["target1"]=df_terror["target1"].fillna("Civilians")
df_terror["motive"]=df_terror["motive"].fillna("Unknown")
df_terror["weaponsub"]=df_terror["weaponsub"].fillna("Unknown Explosive Type")
df_terror["perp"]=df_terror["perp"].fillna(df_terror["perp"].median())
df_terror["killed"]=df_terror["killed"].fillna(df_terror["killed"].median())
df_terror["wounded"]=df_terror["wounded"].fillna(df_terror["wounded"].median())
df_terror["casualities"]=df_terror["casualities"].fillna(df_terror["casualities"].median())
```

```python
df_terror.isnull().sum()
```

### All the null values have been taken care of!
